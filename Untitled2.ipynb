```python
# prompt: now fill the average according to the pollutant id

# Create a dictionary to store the average values for each pollutant ID
pollutant_averages = {}

# Iterate through each pollutant ID
for pollutant_id in tamil_nadu_df['pollutant_id'].unique():
  # Calculate the average pollutant_min value for the current pollutant ID
  average_value = tamil_nadu_df[tamil_nadu_df['pollutant_id'] == pollutant_id]['pollutant_avg'].mean()
  # Store the average value in the dictionary
  pollutant_averages[pollutant_id] = average_value

# Fill the missing values in the 'pollutant_min' column using the dictionary
tamil_nadu_df['pollutant_avg'].fillna(lambda x: pollutant_averages[x['pollutant_id']], inplace=True)
```

```python
# prompt: now using linear regression predict next values of avg pollutant of tamil_nadu_df and create a seperate column and give each predicted values

import numpy as np
from sklearn.linear_model import LinearRegression

# Create a linear regression model
model = LinearRegression()

# Train the model using the average pollutant values from tamil_nadu_df
model.fit(tamil_nadu_df['year'].values.reshape(-1, 1), tamil_nadu_df['avg_pollutant'].values)

# Predict the next values of average pollutant for tamil_nadu_df
predicted_values = model.predict(tamil_nadu_next_df['year'].values.reshape(-1, 1))

# Create a new column in tamil_nadu_next_df for the predicted values
tamil_nadu_next_df['predicted_avg_pollutant'] = predicted_values

# Display the updated tamil_nadu_next_df
print(tamil_nadu_next_df)
```
